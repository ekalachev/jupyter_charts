```python
import numpy as np
import pandas as pd
import time

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# Shows scaling of the sum of two vectors: animation
fig = plt.figure(figsize=(5,5))

def draw_vector(i):
    # clear prior frame    
    plt.clf()
    
    tail = [0, 0]    
    
    basis = np.array([[0, 1], [1, 0]])
    
    basis = basis * (i - 5)
    
    plt.gca().quiver(*tail, *basis[:, 0], angles='xy', scale_units='xy', color='blue', width=0.005, scale=1)
    plt.gca().quiver(*basis[:, 0], *basis[:, 1], angles='xy', scale_units='xy', color='orange', width=0.005, scale=1)
    
    vector = basis.sum(axis=0)
    
    plt.gca().quiver(*tail, *vector, angles='xy', scale_units='xy', color='red', width=0.01, scale=1)
    
    plt.gca().set_xlim([-6, 6])
    plt.gca().set_ylim([-6, 6])
    plt.grid()

anim = animation.FuncAnimation(fig, draw_vector, interval=500, frames=11, repeat=False)
```

```python
# Shows scaling of the sum of two vectors: animation
fig = plt.figure(figsize=(6,6))

def draw_vector(i):
    # clear prior frame    
    plt.clf()
    
    tail = [0, 0]    
    
    basis = np.array([[0, 1], [1, 0]])
    
    if i < 6:  
        basis = basis * np.array([5, i])
    elif i < 16:
        basis = basis * np.array([10-i, 5])
    elif i < 26:
        basis = basis * np.array([-5, 20-i])
    elif i < 36:
        basis = basis * np.array([-30+i, -5])
    else:
        basis = basis * np.array([5, -40+i])
    
    plt.gca().quiver(*tail, *basis[:, 0], angles='xy', scale_units='xy', color='blue', width=0.003, scale=1)
    plt.gca().quiver(*basis[:, 0], *basis[:, 1], angles='xy', scale_units='xy', color='orange', width=0.003, scale=1)
    
    vector = basis.sum(axis=1)
    
    plt.gca().quiver(*tail, *vector, angles='xy', scale_units='xy', color='red', width=0.01, scale=1)
    
    plt.gca().set_xlim([-5.5, 5.5])
    plt.gca().set_ylim([-5.5, 5.5])
    plt.grid()

anim = animation.FuncAnimation(fig, draw_vector, interval=300, frames=41, repeat=False)
```
