```python
import numpy as np
import pandas as pd
import time

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

# Parabola

$$
f(x) = x^2
$$

```python
%matplotlib inline

fig = plt.figure(figsize=(6, 4))

# х - аргумент в нашей функции.
# передадим начало и конец числового отрезка нашей функции и шаг аргумента:
x = np.arange(-10, 10.01, 0.01)

plt.plot(x, x**2)

plt.ylabel('f(x)')
plt.xlabel('x')

plt.title('Parabola: $y = x^{2}$', loc='right')

plt.grid(True)

plt.show()
```

# Exponent

$$
f(x) = e^x
$$

```python
%matplotlib inline

fig = plt.figure(figsize=(6, 4))

x = np.arange(-2, 10, 0.01)

plt.plot(x, np.e**x)

plt.ylabel('f(x)')
plt.xlabel('x')

plt.title('Exponent: $y = e^{x}$', loc='right')

plt.grid(True)

plt.show()
```

# The standard normal probability density function

$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
$$

```python
%matplotlib inline

fig = plt.figure(figsize=(6, 4))

x = np.arange(-6, 6, 0.01)

plt.plot(x, (1/np.sqrt(2*np.pi))*np.e**(-(x**2/2)))

plt.ylabel('f(x)')
plt.xlabel('x')

plt.title('The standard normal probability density function')

plt.grid(True)

plt.show()
```

# Sigmoid

$$
f(x) = \frac{1}{1 + e^{-\alpha x}}
$$

```python
%matplotlib inline

fig = plt.figure(figsize=(6, 4))

x = np.arange(-1, 1, 0.01)

plt.plot(x, (1/(1 + np.e**(-5*x))))
plt.plot(x, (1/(1 + np.e**(-10*x))))

plt.ylabel('f(x)')
plt.xlabel('x')

plt.title('Sigmoid')

plt.grid(True)

plt.show()
```

```python
# Shows scaling of the sum of two vectors: animation
fig = plt.figure(figsize=(5, 5))


def draw_vector(i):
    # clear prior frame
    plt.clf()

    tail = [0, 0]

    basis = np.array([[0, 1], [1, 0]])

    basis = basis * (i - 5)

    plt.gca().quiver(*tail, *basis[:, 0], angles='xy',
                     scale_units='xy', color='blue', width=0.005, scale=1)
    plt.gca().quiver(*basis[:, 0], *basis[:, 1], angles='xy',
                     scale_units='xy', color='orange', width=0.005, scale=1)

    vector = basis.sum(axis=0)

    plt.gca().quiver(*tail, *vector, angles='xy',
                     scale_units='xy', color='red', width=0.01, scale=1)

    plt.gca().set_xlim([-6, 6])
    plt.gca().set_ylim([-6, 6])
    plt.grid()


anim = animation.FuncAnimation(
    fig, draw_vector, interval=200, frames=11, repeat=False)
```

```python
# Shows scaling of the sum of two vectors: animation
fig = plt.figure(figsize=(5, 5))


def draw_vector(i):
    # clear prior frame
    plt.clf()

    tail = [0, 0]

    basis = np.matrix([[0, 1], [1, 0]])

    if i < 6:
        vector = basis.dot(np.array([5, i]))
    elif i < 16:
        vector = basis.dot(np.array([10-i, 5]))
    elif i < 26:
        vector = basis.dot(np.array([-5, 20-i]))
    elif i < 36:
        vector = basis.dot(np.array([-30+i, -5]))
    else:
        vector = basis.dot(np.array([5, -40+i]))

    plt.gca().quiver(*tail, *vector.A1, angles='xy',
                     scale_units='xy', color='red', width=0.01, scale=1)

    plt.gca().set_xlim([-5.5, 5.5])
    plt.gca().set_ylim([-5.5, 5.5])
    plt.grid()


anim = animation.FuncAnimation(
    fig, draw_vector, interval=100, frames=41, repeat=False)
```
