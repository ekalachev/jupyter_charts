```python
import numpy as np
import pandas as pd
import time

import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

# Parabola

$$
f(x) = x^2
$$

```python
# Set plot size
fig = plt.figure(figsize=(8, 6))

plt.title('Parabola: $y = x^{2}$', loc='right')
plt.ylabel('f(x)')
plt.xlabel('x')

plt.grid(True)

# х - аргумент в нашей функции.
# передадим начало и конец числового отрезка нашей функции и шаг аргумента:
x = np.arange(-10, 10.01, 0.01)

plt.plot(x, x**2)

plt.show()
```

# Exponent

$$
f(x) = e^x
$$

```python
# Set plot size
fig = plt.figure(figsize=(8, 6))

plt.ylabel('f(x)')
plt.xlabel('x')
plt.title('Exponent: $y = e^{x}$', loc='right')

plt.grid(True)

x = np.arange(-2, 10, 0.01)
plt.plot(x, np.e**x)

plt.show()
```

# The standard normal probability density function

$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
$$

```python
# Set plot size
fig = plt.figure(figsize=(8, 6))

plt.title('The standard normal probability density function')
plt.ylabel('f(x)')
plt.xlabel('x')

plt.grid(True)

x = np.arange(-6, 6, 0.01)
plt.plot(x, (1/np.sqrt(2*np.pi))*np.e**(-(x**2/2)))

plt.show()
```

# Sigmoid

$$
f(x) = \frac{1}{1 + e^{-\alpha x}}
$$

```python
# Set plot size
fig = plt.figure(figsize=(8, 6))

plt.title('Sigmoid')
plt.legend(['alpha = -5', 'alpha = -10'])
plt.ylabel('f(x)')
plt.xlabel('x')

plt.grid(True)

x = np.arange(-1, 1, 0.01)

plt.plot(x, (1/(1 + np.e**(-5*x))))
plt.plot(x, (1/(1 + np.e**(-10*x))))

plt.show()
```
