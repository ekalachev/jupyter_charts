```python
import numpy as np
import pandas as pd
import time

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
# Important for animation
%matplotlib notebook
```

```python
# Shows scaling of the sum of two vectors: animation
fig = plt.figure(figsize=(5,5))

def draw_vector(i):
    plt.clf() # ÑÑ‚Ð¸Ñ€Ð°ÑŽ Ð¿Ñ€ÐµÐ´Ñ‹Ð´ÑƒÑˆÐ¸Ð¹ ÐºÐ°Ð´Ñ€
    
    tail = [0, 0]    
    
    b1 = np.array([0, 1])
    b2 = np.array([1, 0])
    
    b1 = b1 * (i - 5)
    b2 = b2 * (i - 5)
    
    v = b1 + b2
    
    plt.gca().quiver(*tail, *b1, angles='xy', scale_units='xy', color='blue', width=0.005, scale=1)
    plt.gca().quiver(*b1, *b2, angles='xy', scale_units='xy', color='orange', width=0.005, scale=1)
    plt.gca().quiver(*tail, *v, angles='xy', scale_units='xy', color='red', width=0.01, scale=1)
    
    plt.gca().set_xlim([-6, 6])
    plt.gca().set_ylim([-6, 6])
    plt.grid()

anim = animation.FuncAnimation(fig, draw_vector, interval=500, frames=11, repeat=False)
```
