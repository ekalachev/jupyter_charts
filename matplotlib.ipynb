# Parabola

$$
f(x) = x^2
$$

```python
# Set plot size
fig = plt.figure(figsize=(10, 6))

plt.title(r'Parabola: $y = x^{2}$', fontsize=18, loc='right')
plt.ylabel('f(x)', fontsize=14)
plt.xlabel('x', fontsize=14)

plt.grid(True)
plt.axhline(0, color='black')
plt.axvline(0, color='black')

# х - аргумент в нашей функции.
# передадим начало и конец числового отрезка нашей функции и шаг аргумента:
x = np.arange(-10, 10.01, 0.01)

plt.plot(x, x**2)

plt.show()
```

# The standard normal probability density function

$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
$$

```python
def get_standard_normal_probability_density_function(x):
    return (1/np.sqrt(2 * np.pi)) * np.e**(-(x**2 / 2))


# Set plot size
fig = plt.figure(figsize=(12, 6))

plt.title(
    r'The standard normal probability density function: $f(x) = \dfrac{1}{\sqrt{2\pi}} e^{-\dfrac{x^2}{2}}$',
    fontsize=18
)
plt.ylabel('f(x)', fontsize=14)
plt.xlabel('x', fontsize=14)

plt.grid(True)

x = np.arange(-6, 6, 0.01)
plt.plot(x, get_standard_normal_probability_density_function(x))

plt.show()
```

# Sigmoid

$$
f(x) = \frac{1}{1 + e^{-\alpha x}}
$$

```python
def get_sigmoid(x, alpha):
    return 1/(1 + np.e**(-alpha * x))


# Set plot size
fig = plt.figure(figsize=(10, 6))

plt.title(r'Sigmoid: $f(x) = \dfrac{1}{1 + e^{-\alpha x}}$', fontsize=18, loc='right')
plt.ylabel('f(x)', fontsize=14)
plt.xlabel('x', fontsize=14)

plt.grid(True)
plt.axhline(0, color='blue', lw=1.2)
plt.axvline(0, color='blue', lw=1.2)

x = np.arange(-1, 1, 0.01)

plt.plot(x, get_sigmoid(x, 5))
plt.plot(x, get_sigmoid(x, 10))

plt.legend([
    r'$f(x) = \dfrac{1}{1 + e^{-5 x}}$',
    r'$f(x) = \dfrac{1}{1 + e^{-10 x}}$'
])

plt.show()
```
