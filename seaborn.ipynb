```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

# Categorical charts

```python
df = pd.DataFrame({
    'number_of_attempts': [1, 2, 3, 4, 5, 6, 7, 8],
    'probability': [0.05, 0.1, 0.3, 0.25, 0.15, 0.07, 0.05, 0.03]
})

# size of chart
plt.rcParams['figure.figsize'] = (12, 6)
# style of grid (darkgrid, whitegrid, dark, white, ticks)
sns.set_style('whitegrid')

chart = sns.barplot(data=df, x='number_of_attempts', y='probability', palette='Blues_d')

chart.set_title('Probability distribution X')
chart.set_xlabel('Number of attempts')
chart.set_ylabel('Probability')

plt.show()
```
