```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

# Categorical charts

```python
df = pd.DataFrame({
    'number_of_attempts': [1, 2, 3, 4, 5, 6, 7, 8],
    'probability': [0.05, 0.1, 0.3, 0.25, 0.15, 0.07, 0.05, 0.03]
})

# size of chart
plt.rcParams['figure.figsize'] = (12, 6)
# style of grid (darkgrid, whitegrid, dark, white, ticks)
sns.set_style('whitegrid')

chart = sns.barplot(data=df, x='number_of_attempts', y='probability', palette='Blues_d')

chart.set_title('Probability distribution X', fontsize=25)
chart.set_xlabel('Number of attempts', fontsize=15)
chart.set_ylabel('Probability', fontsize=15)

plt.show()
```

# Histogram

A histogram is a plot that lets you discover, and show, the underlying frequency distribution (shape) of a set of continuous data. 

```python
hist_df = pd.DataFrame({
    'ranking': [100, 120, 290, 210, 395, 199, 505,105, 407]
})
```

```python
# bins: how many bars
# kde: compute a kernel density estimate to smooth the distribution and show on the plot as (one or more) line(s). 
# Only relevant with univariate data.
chart = sns.histplot(data=hist_df['ranking'], bins=5, kde=True)

chart.set_title('Histogram of ranking', fontsize=25)
chart.set_xlabel('Ranking', fontsize=15)
chart.set_ylabel('Number of observations', fontsize=15)
chart.tick_params(labelsize=10)
```
